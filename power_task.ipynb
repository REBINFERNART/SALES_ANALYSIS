```python
import pandas as pd

# Load datasets
orders = pd.read_csv(r"C:\Users\krebi\Downloads\orders.csv")
order_details = pd.read_csv(r"C:\Users\krebi\Downloads\order_details.csv")
customers = pd.read_csv(r"C:\Users\krebi\Downloads\customers_decode.csv")
employees = pd.read_csv(r"C:\Users\krebi\Downloads\employees.csv")
shippers = pd.read_csv(r"C:\Users\krebi\Downloads\shippers.csv")
products = pd.read_csv(r"C:\Users\krebi\Downloads\products_decode.csv")
categories = pd.read_csv(r"C:\Users\krebi\Downloads\categories.csv")

# Merge datasets using 'left' join to keep all orders
combined = pd.merge(orders, order_details, on='orderID', how='left')
combined = pd.merge(combined, customers, on='customerID', how='left')
combined = pd.merge(combined, employees, on='employeeID', how='left')
combined = pd.merge(combined, shippers, on='shipperID', how='left')
combined = pd.merge(combined, products, on='productID', how='left')
combined = pd.merge(combined, categories, on='categoryID', how='left')

# Rename columns to remove '_x' and '_y' conflicts
combined.rename(columns={
    'companyName_x': 'customerCompany',
    'companyName_y': 'supplierCompany',
    'city_x': 'customerCity',
    'city_y': 'supplierCity',
    'country_x': 'customerCountry',
    'country_y': 'supplierCountry',
    'unitPrice_x': 'orderUnitPrice',
    'unitPrice_y': 'productUnitPrice'
}, inplace=True)

# Select relevant columns with updated names
final_columns = [
    'orderID', 'customerID', 'employeeID', 'orderDate', 'requiredDate', 'shippedDate', 
    'shipperID', 'freight', 'customerCompany', 'contactName', 'contactTitle', 
    'customerCity', 'customerCountry', 'employeeName', 'title', 'reportsTo', 
    'supplierCompany', 'productID', 'productName', 'quantityPerUnit', 'orderUnitPrice', 
    'discontinued', 'categoryID', 'categoryName', 'description', 'quantity', 'discount'
]

# Ensure all selected columns exist in the dataset
final_columns = [col for col in final_columns if col in combined.columns]

# Save the final dataset
combined[final_columns].to_csv('combined_dataset.csv', index=False)

print("Dataset successfully saved as 'combined_dataset.csv'")
```
